Calculating number of syllables:

- if the word is less than or equal to three letters long, it has one syllable.
- count the vowels, that's your syllable count.
- worry about 'aa', 'ee', 'ii', 'oo', 'uu', then decrement if found.
- if the last letter is 'e', decrement.
- if the last letter is 'y', increment.

```python
bad_word = 'loaded'
```

```python
def how_many_syllables(word):
    if len(word) <= 3:
        return 1
    nv = 0
    for w in word:
        if w in 'aeiou':
            nv += 1
    return nv
```

```python
def how_many_syllables(word):
    if len(word) <= 3:
        return 1
    nv = 0
    for w in word:
        if w in 'aeiou':
            nv += 1
    for pair in ['aa', 'ee', 'ii', 'oo', 'uu']:
        if word.find(pair) != -1:
            nv -= 1
    return nv
```

```python
def how_many_syllables(word):
    if len(word) <= 3:
        return 1
    nv = 0
    for w in word:
        if w in 'aeiou':
            nv += 1
    for pair in ['aa', 'ee', 'ii', 'oo', 'uu']:
        if word.find(pair) != -1:
            nv -= 1
    if word[-1] == 'e':
        nv -= 1
    if word[-1] == 'y':
        nv += 1
    return nv
```
