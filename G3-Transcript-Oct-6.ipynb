```python
def how_many_syllables(word):
    """   ???  """
    if len(word) <= 3:
        return 1
    nv = 0
    for w in word:
        if w in 'aeiou':
            nv += 1
    for pair in ['aa', 'ee', 'ii', 'oo', 'uu']:
        if word.find(pair) != -1:
            nv -= 1
    if word[-1] == 'e':
        nv -= 1
    if word[-1] == 'y':
        nv += 1
    return nv
```

```python
def calc_reading_level(filename):
    """ ??? """
    with open(filename, errors='ignore') as df:
        rawdata = df.read()
    words = rawdata.split()
    total_words = len(words)
    
    nstops = rawdata.count('.')
    nquestions = rawdata.count('?')
    nexclaims = rawdata.count('!')
    nsemis  = rawdata.count(';')
    ncolons = rawdata.count(':')   
    
    total_sentences = nstops + nquestions + nexclaims + nsemis + ncolons
    
    total_syllables = 0
    for word in words:
        total_syllables += how_many_syllables(word)
    
    return 206.835 - (1.015 * (total_words/total_sentences)) - (84.6 * (total_syllables/total_words))
    
```
